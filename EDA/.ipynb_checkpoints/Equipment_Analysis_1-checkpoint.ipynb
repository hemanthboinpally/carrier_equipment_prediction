## Topic Modeling - Using Commodity

Based on my business understanding , the commodity type of the shipment would certainly play a par for the type of equipment.

```python
fmcsa_df["Commodity type Classification"].value_counts()
```

Columns looks to be clean, So I didn't do any data cleansing - stopwards, stemming etc

```python
import nltk
from nltk import word_tokenize
import re
```

```python
commodity = fmcsa_df["Commodity type Classification"].fillna('NA')
```

```python
len(commodity)
```

```python
from gensim import corpora
```

```python
cleaned_tokens = commodity.apply(lambda row: word_tokenize(row))

# No. of times a word appears in the entire pool of documents
dictionary = corpora.Dictionary(cleaned_tokens)

# Filter out (non) frequent words
dictionary.filter_extremes(no_below=5, keep_n=5000)

corpus = [dictionary.doc2bow(word) for word in cleaned_tokens]
```

```python
import gensim
ldamodel = gensim.models.ldamodel.LdaModel(corpus,num_topics=3,id2word=dictionary,passes=15)
```

```python
topics = ldamodel.print_topics(num_words=4)
```

```python
for topic in topics:
    print(topic)
```

```python
import pyLDAvis.gensim
lda_display = pyLDAvis.gensim.prepare(ldamodel,corpus,dictionary,sort_topics=False)
```

```python
pyLDAvis.display(lda_display)
```

```python
def get_topic_details(ldamodel, corpus):
    topic_details_df = pd.DataFrame()
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_details_df = topic_details_df.append(pd.Series([topic_num, prop_topic]), ignore_index=True)
    topic_details_df.columns = ['Dominant_Topic', '% Score']
    return topic_details_df
```

```python
topic_score_non_split = get_topic_details(ldamodel, corpus)
```

```python
clustered_df_1 = topic_score_non_split.join(fmcsa_df)
```

```python
clustered_df_1.sort_values(by=['Dominant_Topic'])
```

```python
clustered_df_1.groupby(['Dominant_Topic','Commodity type Classification']).size()
```

<b> Observation: </b> Seems to be nicely grouped together but, since some commodities should have been grouped together like meat, refrigrated. I will try some other technique

##  K-Means Model

```python
# Perform the necessary imports
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
```

```python
# Create scaler: scaler
scaler = StandardScaler()
```

```python
scaler.fit(label_encoded_df)
```

```python
new_data = scaler.transform(label_encoded_df)
```

```python
ks = range(1, 18)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(new_data)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

<b> Observation : </b> k = 13 seems like a good number to choose for clustering 

## Pipeline for Production Implementation

```python
# Create scaler: scaler
scaler = StandardScaler()

# Create KMeans instance: kmeans
kmeans = KMeans(n_clusters=13)

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,kmeans)
```

```python
# Fit the pipeline to samples
pipeline.fit(label_encoded_df)
```

```python
# Calculate the cluster labels: labels
labels = pipeline.predict(label_encoded_df)
```

```python
labels
```

```python
label_encoded_kmeans = label_encoded_df.copy()
```

```python
label_encoded_kmeans['Prediction'] = labels
```
